```python
import pandas as pd
import re
```

```python
f = open("WhatsApp Chat with Hostel Mates.txt","r",encoding = "utf-8")
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[AP]M\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
```

```python
messages
```

```python
dates1 = re.findall(pattern, data)
```

```python
dates1
```

```python
pattern1 = r'\u202f[AP]M'
```

```python
cleaned_dates = [re.sub(pattern1, '', date) for date in dates1]
```

```python
cleaned_dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': cleaned_dates})
```

```python
df.head()
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %H:%M - ')
df.rename(columns={'message_date': 'date'}, inplace=True)
```
